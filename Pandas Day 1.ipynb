```python
grades = pd.read_excel('student.xlsx',sheet_name='tblgrade')
```

```python
first_join = pd.merge(left=usermaster, right= users, left_on='usermasterid', right_on='usermasterid', how = 'inner')
```

```python
first_join = first_join[['usermasterid', 'tbluser', 'email']]
```

```python
second_join = pd.merge(left=first_join, right= userdetails[['firstname', 'userid']], left_on='tbluser', right_on='userid', how = 'inner')
```

```python
second_join.head()
```

```python
third_join = pd.merge(second_join, usergrade[['userdetailid', 'gradeid']], left_on='userid', right_on='userdetailid', how = 'inner')
```

```python
third_join.head()
```

```python
#if join column name is same the use single on else use left_on, right_on etc
#join with selecting desired columns only from new table


last_join = pd.merge(third_join, grades[['gradeid', 'gradename']], on = 'gradeid', how='inner')
```

```python
last_join.head(1)
```

```python
#excluding unwanted columns

df = last_join[last_join.columns[~last_join.columns.isin(['tbluser','userdetailid','gradeid'])]]
```

```python
#arranging column sequence

df = df[['usermasterid', 'userid', 'firstname', 'gradename', 'email']]
```
